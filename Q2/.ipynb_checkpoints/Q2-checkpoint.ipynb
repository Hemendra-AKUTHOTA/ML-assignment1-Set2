### Import required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
plt.style.use('dark_background')
```

### Load the data file

```python
census_income = pd.read_csv('adult.csv')
census_income.head()
```

**Dropping the below columns:**
> 1. education.num as education is already giving similar insights <br>
> 2. fnlwgt is highly discrete <br>

> 4. capital.gain is highly discrete <br>

```python
# census_income = census_income.drop(['education', 'fnlwgt', 'capital.loss', 'capital.gain'], axis = 1)
census_income = census_income.drop(['education', 'fnlwgt'], axis = 1)
census_income.head(1)
```

### Replacing ? with NaN

```python
census_income.replace('?', np.NaN,inplace = True)
census_income.head()
```

### Replacing NaN with mode 

```python
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
census_income = pd.DataFrame(imp.fit_transform(census_income),columns=census_income.columns,index=census_income.index)
```

```python
census_income.isnull().sum()
# census_income.info()
```

### Label Encoding

```python
le = LabelEncoder()
census_income['age'] = le.fit_transform(census_income['age'])
census_income['workclass'] = le.fit_transform(census_income['workclass'])
census_income['education.num'] = le.fit_transform(census_income['education.num'])
census_income['marital.status'] = le.fit_transform(census_income['marital.status'])
census_income['occupation'] = le.fit_transform(census_income['occupation'])
census_income['relationship'] = le.fit_transform(census_income['relationship'])
census_income['race'] = le.fit_transform(census_income['race'])
census_income['sex'] = le.fit_transform(census_income['sex'])
census_income['capital.gain'] = le.fit_transform(census_income['capital.gain'])
census_income['capital.loss'] = le.fit_transform(census_income['capital.loss'])
census_income['hours.per.week'] = le.fit_transform(census_income['hours.per.week'])
census_income['native.country'] = le.fit_transform(census_income['native.country'])
census_income['income'] = le.fit_transform(census_income['income'])

census_income.head()
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(census_income.corr(),annot=True)
```

### Drop the columns that has correlation between 0.1 and -0.1

```python
census_income = census_income.drop(['workclass', 'occupation','race','native.country'], axis = 1)
```

```python
y = census_income['income']
x = census_income.drop(['income'], axis = 1)
```

# **Train Test Split**

```python

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2) 
```

### Gaussian Naive Bayes

```python
gb = GaussianNB()
gb.fit(x_train,y_train)
```

### Printing Metrics

```python
y_pred = gb.predict(x_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred)*100)
```

### Bernouli Naive Bayes

```python
bnb = BernoulliNB()
bnb.fit(x_train,y_train)
```

### Printing Metrics

```python
y_pred = bnb.predict(x_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred)*100)
```

### Multinomial Naive Bayes

```python
mnb = MultinomialNB()
mnb.fit(x_train,y_train)
```

### Printing Metrics

```python
y_pred = mnb.predict(x_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred)*100)
```
