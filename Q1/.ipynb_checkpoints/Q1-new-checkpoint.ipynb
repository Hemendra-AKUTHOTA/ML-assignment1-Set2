# Classify cars data on acceptibility

### Import required libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
```

## Exploratory analysis 

### 1. Load the data file
The data need to be converted into the below format<br>
buying, maint, doors, persons, lug_boot, safety, cond_class <br>
[vhigh,low,4,more,small,med,unacc]

```python
car_data = pd.read_csv('car.data',sep=',')
car_data.columns = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'cond_class']
```

### 2. Check Data anomalies
There are some data anomalies in 'doors' and 'persons' columns, hence replace "5more" string with 5 in 'doors' column <br>
and "more" string with 6 value

```python
#Analyze the data to see if there is any data cleaning required
# car_data.groupby(['buying']).size()
# car_data.groupby(['maint']).size()
# car_data.groupby(['doors']).size() --- data cleaning is needed
# car_data.groupby(['persons']).size() --- data cleaning is needed
# car_data.groupby(['lug_boot']).size()
# car_data.groupby(['safety']).size()
# car_data.groupby(['cond_class']).size()
# car_data.info() - no null values
# car_data.groupby(['cond_class']).size()
```

```python
car_data['doors'] = car_data['doors'].replace("5more",'5')
car_data['persons'] = car_data['persons'].replace("more",'6')
```

```python
#Split the data into X columns and y column
car_data_X = car_data[['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety']]
car_data_y = car_data[['cond_class']]
```

```python
car_data_buying = pd.crosstab(car_data.cond_class,car_data.buying)
car_data_maint = pd.crosstab(car_data.cond_class,car_data.maint)
car_data_doors = pd.crosstab(car_data.cond_class,car_data.doors)
car_data_persons = pd.crosstab(car_data.cond_class,car_data.persons)
car_data_lug_boot = pd.crosstab(car_data.cond_class,car_data.lug_boot)
car_data_safety = pd.crosstab(car_data.cond_class,car_data.safety)
```

```python
#Take the columns with correlations more than 0.2 and less than -0.2
# car_data_new = car_data_dummies[['persons_2','persons_4','persons_6','safety_high','safety_low','cond_class_acc']]
car_data_new = car_data_dummies[['persons_2','safety_high','safety_low','cond_class_acc']]
```

```python
car_data_new.head()
```

```python
#Dividing the dataframe into x features and y target variable
x = car_data_new.iloc[:, :-1]
y = car_data_new.iloc[:, 3]
```

### Extract X as all columns except the last column and Y as last column

```python
# x.columns = ['persons_2','persons_4','persons_6','safety_high','safety_low']
x.columns = ['persons_2','safety_high','safety_low']
y.columns=['cond_class_acc']
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from numpy import mean
from numpy import std

cv = KFold(n_splits=10, random_state=1, shuffle=True)
# create model
model = LogisticRegression()
# evaluate model
scores = cross_val_score(model, x, y, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
position = 1
for x in range(len(scores)): 
    print('Model#',position,' - ', scores[x])
    position = position+1
print('Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```
